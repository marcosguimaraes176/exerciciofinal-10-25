```python
def tratar_dados_musica_rede(df):
    # Renomear a primeira coluna
    df = df.rename(columns={'TOTAL DE MUNIC√çPIOS': 'MUNIC√çPIO'})
    
    # Lista dos projetos
    projetos = ['VIOL√ïES', 'CORAIS', 'BANDAS', 'ORQUESTRA']
    
    # Criar coluna indicadora dos projetos presentes
    def obter_projetos_presentes(row):
        projetos_presentes = []
        for projeto in projetos:
            if pd.notna(row[projeto]) and row[projeto] != '':
                projetos_presentes.append(projeto)
        return ', '.join(projetos_presentes) if projetos_presentes else 'NENHUM'
    
    df['PROJETOS_PRESENTES'] = df.apply(obter_projetos_presentes, axis=1)
    
    # Filtrar apenas munic√≠pios com pelo menos um projeto
    df_filtrado = df[df['PROJETOS_PRESENTES'] != 'NENHUM'].copy()
    
    # Resetar √≠ndice
    df_filtrado = df_filtrado.reset_index(drop=True)
    
    print(f"‚úÖ Dados tratados: {len(df_filtrado)} munic√≠pios com projetos")
        
    # SALVAR ANTES DE RETORNAR
    df_filtrado.to_excel('projetos_tratados.xlsx', index=False)
    print("üíæ Arquivo 'projetos_tratados.xlsx' salvo com sucesso!")
        
    return df_filtrado

# üöÄ EXECUTAR A FUN√á√ÉO

df_tratado = tratar_dados_musica_rede(projetos_por_municipios)

if df_tratado is not None:
    print("\nüìä Resumo:")
    print(f"Total de munic√≠pios com projetos: {len(df_tratado)}")
    print("\nPrimeiras linhas:")
    print(df_tratado[['MUNIC√çPIO', 'PROJETOS_PRESENTES']].head())
else:
    print("Falha ao processar os dados.")
```

```python
# Lista dos projetos
projetos = ['VIOL√ïES', 'CORAIS', 'BANDAS', 'ORQUESTRA']

# Dicion√°rio para armazenar os dataframes de cada projeto
dfs_por_projeto = {}


for projeto in projetos:
   
    # Seleciona apenas os munic√≠pios que possuem o projeto atual
    df_projeto = df_tratado[df_tratado[projeto].notna() & (df_tratado[projeto] != '')].copy()
    # Mant√©m apenas as colunas relevantes
    df_projeto = df_projeto[['MUNIC√çPIO', projeto]].rename(columns={projeto: 'DETALHE_PROJETO'})
    dfs_por_projeto[projeto] = df_projeto

# Agora, por exemplo, dfs_por_projeto['VIOL√ïES'] cont√©m os munic√≠pios com apenas o projeto VIOL√ïES
# Para acessar os dataframes:
df_violao = dfs_por_projeto['VIOL√ïES']
df_corais = dfs_por_projeto['CORAIS']
df_bandas = dfs_por_projeto['BANDAS']
df_orquestra = dfs_por_projeto['ORQUESTRA']

# Exemplo de visualiza√ß√£o
df_violao.head()
df_corais.head()
df_bandas.head()
df_orquestra.head()
```
